### enumerate

```python
# enumerate function gets index and its value and stores it in a tuple
a = ['eat', 'sleep', 'repeat']
ans = enumerate(a)

print(list(ans))
```

```python
# enumerate using in a string example
a = "Abbas"
b = enumerate(a, 10)   # second argument tells where to start indexing

print(list(b))
```

```python
data.columns
```

```python
# To remove any column
categorical_values = list(data.columns)
categorical_values.remove('target')
categorical_values
```

```python
print(list(enumerate(categorical_values, 1)))
```

```python
print(categorical_values)
```

```python
plt.figure(figsize=(30,30))

for i, col in enumerate(categorical_values, 1):
    plt.subplot(4,4,i)   # 4/4 dimension     total 16 plots         and plot on i's value
    sns.barplot(x=f"{col}", y='target', data=data)
    plt.ylabel("Possibility of having heart disease")
    plt.xlabel(f"{col}")
```
